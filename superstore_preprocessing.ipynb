```python
# Step 1: Install and import necessary libraries
import pandas as pd

# Step 2: Upload the dataset (you'll be prompted to upload Superstore.csv)
from google.colab import files
uploaded = files.upload()

# Step 3: Read the CSV file
df = pd.read_csv('SampleSuperstore.csv', encoding='latin1')  # Replace with your filename if different

# Step 4: View first few rows
df.head()

# Step 5: Convert 'Order Date' to datetime format and handle errors
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Step 6: Remove rows with invalid or missing 'Order Date'
df = df.dropna(subset=['Order Date'])

# Step 7: Optional: Also clean other key columns (e.g., Sales, Profit)
# Convert columns to numeric (if not already)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')

# Drop rows with nulls in key financial columns
df = df.dropna(subset=['Sales', 'Profit', 'Discount'])

# Step 8: Add new derived columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()
df['Profit Ratio'] = df['Profit'] / df['Sales']

# Step 9: View cleaned data
df.head()

# Step 10: Export cleaned data as CSV
cleaned_file = 'Superstore_Cleaned.csv'
df.to_csv(cleaned_file, index=False)

# Step 11: Download the cleaned CSV
files.download(cleaned_file)
```
